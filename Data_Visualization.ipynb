````python
from pathlib import Path
import tarfile
import urllib.request
import pandas as pd
import os

# Define a function to load the housing data
def load_voting_data():
    # Define the path to the tarball file that contains the dataset
    tarball_path = Path("datasets/voterdata.tgz")
        # Check if the tarball file already exists locally
    if not tarball_path.is_file():
        # If the file doesn't exist, create the 'datasets' directory (if it doesn't exist)
        Path("datasets").mkdir(parents=True, exist_ok=True)

        # URL of the tarball file containing the dataset
        url = "https://github.com/Momshroom/ML1Proj5DataTeamNBC/raw/refs/heads/main/voterdata.tgz"

        # Download the tarball file from the URL and save it to the specified path
        urllib.request.urlretrieve(url, tarball_path)

        # Open the tarball file for extraction
        with tarfile.open(tarball_path) as voterdata_tarball:
            # Extract all the files in the tarball to the 'datasets' directory
            voterdata_tarball.extractall(path="datasets")

    # Load the housing dataset (CSV) from the extracted files and return it as a pandas DataFrame
    return pd.read_csv(Path("datasets/Project5TeamNBC.csv")),pd.read_csv(Path("datasets/Project5_CSVFormat_Presidential.csv")),pd.read_csv(Path("datasets/Project5_CSVFormat_Offcycle.csv"))

# Call the function to load the housing data and store it in the 'housing' variable
voting, presidential, offcycle = load_voting_data()
````

#**Voting Trends by Ethnicity and Age**

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Used to select which ethnicity to view
groups = ['White', 'White Non-Hispanic', 'Black', 'Asian', 'Hispanic']
ethnicity = groups[2]

# Option to save generated file
save_file = False


def plot_columns_from_multiple_files(file_paths, column_name):
    idx = 0

    plt.figure(figsize=(14, 7))

    # Iterate through the files and plot the specified column
    for i, file_path in enumerate(file_paths):
        # Read the data
        data = pd.read_csv(file_path)
        sorted_data = data.sort_values(by='Year')

        # Normalize the years to start at 0
        normalized_years = sorted_data['Year'] - 2000
        x = normalized_years
        y = sorted_data[column_name]

        age_groups = ['18-24', '25-44', '45-64', '65+']

        # Plot the data for the current file
        line, = plt.plot(x, y, marker='o', label=age_groups[idx])

        idx += 1

        # Fit a regression line
        coef = np.polyfit(x, y, 1)  # Linear fit
        poly1d_fn = np.poly1d(coef)
        plt.plot(x, poly1d_fn(x), linestyle='--', color=line.get_color())

    # Add labels, legend, and title
    plt.xlabel('Year', fontsize=14)
    plt.ylabel('Percentage of Population', fontsize=14)
    plt.title(f'Data for {column_name} Voters by Age', fontsize=16, fontweight='bold')

    # Define the explicit range for x-axis ticks
    tick_positions = np.arange(4, max(normalized_years) + 1, 4)
    tick_labels = tick_positions + 2000
    plt.xticks(ticks=tick_positions, labels=tick_labels, fontsize=12, rotation=45)

    # Add grid with a lighter style
    plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

    # Improve the legend's appearance
    plt.legend(fontsize=12, loc='upper left', bbox_to_anchor=(1.01, 1.016), frameon=True)

    # Show the plot
    plt.tight_layout()
    plt.show()

    if save_file:
        filename = input('Enter Filename: ')
        plt.savefig(f'{filename}.png', dpi=72, bbox_inches='tight')

# Example usage
file_paths = [
    'datasets/CENSUS_1824.csv',
    'datasets/CENSUS_2544.csv',
    'datasets/CENSUS_4564.csv',
    'datasets/CENSUS_65OVER.csv'
]

# Call the function to plot a specific column from all files together
plot_columns_from_multiple_files(file_paths, column_name=ethnicity)
````

#**Pie Chart for Voter Turnout by Year**

````python
import pandas as pd
import matplotlib.pyplot as plt

# True to save file
save_file = False


# Define a function to plot pie charts for each year
def plot_voting_pie_charts(file_path, column):
    colors = ['#90EE90', '#FFB6C1']

    # Read the dataset
    data = pd.read_csv(file_path)

    curr_year = ''

    # Iterate through each year in the dataset
    for _, row in data.iterrows():
        year = row['Year']
        voted_percentage = row[column]
        not_voted_percentage = 1 - voted_percentage

        curr_year = row['Year']

        # Data for the pie chart
        labels = ['Voted', 'Did Not Vote']
        sizes = [voted_percentage, not_voted_percentage]

        # Create the pie chart
        plt.figure(figsize=(6, 6))
        plt.pie(
            sizes,
            labels=labels,
            autopct='%1.1f%%',
            startangle=90,
            colors=colors
        )
        plt.title(f'Voting Distribution in {int(year)}', fontsize=14, fontweight='bold')
        plt.tight_layout()

        plt.savefig(f'pie_chart_{year}.png', format='png', dpi=300)

        plt.show()

        if save_file:
            filename = input('Enter Filename: ')
            plt.savefig(f'{filename}_{curr_year}.png', dpi=72, bbox_inches='tight')

# File path and column to use
file_path = 'datasets/CENSUS_TOTAL.csv'
column = 'Total'  # Column representing the percentage who voted

# Call the function to plot pie charts
plot_voting_pie_charts(file_path, column)
````

````python
# Now to categorize elections based on type of election and create a new column to hold the categories

# First, create a mapping function:
presidential_years = ['2000', '2004', '2008', '2012', '2016', '2020']
offcycle_years = ['2002','2006', '2010', '2014', '2018', '2022']
def categorize_election_type(year):
    if year in presidential_years:
        return "presidential"
    elif year in offcycle_years:
        return "offcycle"
    else:
        return "other"

# now filter and create the new column
df_voters_and_not['election_type'] = df_voters_and_not['YEAR'].apply(categorize_election_type)

# Show that column is there:
df_voters_and_not.head()
````

````python
# Now process the race data into fewer categories:
# define category group lists:
# We will use: White, Black, Native American, AAPI, Other, Multiracial
# some of those have multiple encodings, so defining encoding lists for those:
# multiracial_id = anything starting with 8
# AAPI = anything starting with 6
# Other = 999,700

# make a mapping function:
def categorize_race(race_code):
    if race_code == '100':
        return "White"
    elif race_code == '200':
        return "Black"
    elif race_code == '300':
        return "NativeAmerican"
    elif race_code[0]=='6':
        return "AAPI"
    elif race_code[0]=='8':
        return "Multiracial"
    else:
        return "Other"

# filter and create new column
df_voters_and_not['race_category'] = df_voters_and_not['RACE'].apply(categorize_race)

# Show that column is there:
df_voters_and_not.head()
````

````python
# Create a function that assigns each age to its corresponding age group and ensure that minors under 18 are excluded:

def classify_age_group(age):
    if 18 <= age <= 24:
        return '18_24'
    elif 25 <= age <= 44:
        return '25_44'
    elif 45 <= age <= 65:
        return '45_65'
    elif age > 65:
        return '65over'
    else:
        return 'Under18'

# Apply the function to create a new 'Age_Group' column
df_voters_and_not['Age_Group'] = df_voters_and_not['AGE'].apply(classify_age_group)

# Check the distribution of age groups
print(df_voters_and_not['Age_Group'].value_counts())


````

````python
# Define a function to map EDUC codes to new categories
def recode_education(educ_code):
    educ_code = int(educ_code)
    if 10 <= educ_code <= 32:
        return 'less9th'
    elif 40 <= educ_code <= 72:
        return '9th-12th'
    elif educ_code == 73:
        return 'HSGrad'
    elif 80 <= educ_code <= 92:
        return 'SomeColl'
    elif educ_code == 111:
        return 'Bachelor'
    elif 123 <= educ_code <= 125:
        return 'Advanced'
    else:
        return 'Unknown'

# Apply the function to create a new column
df_voters_and_not['Education_Level'] = df_voters_and_not['EDUC'].apply(recode_education)

# show that it worked:
df_voters_and_not.head()
````

901 is don't know and 902 is N/A or no response.

````python
# Make a new column that is categorical and has Yes, No, and Unknown labels:
# define map function
def map_hispanic(value):
    if value == 0:
        return 'no'
    elif value == 1:
        return 'yes'
    else:
        return 'other'

# generate new column with values:
df_voters_and_not['hispanic_cat'] = df_voters_and_not['HISPAN_BINARY'].apply(map_hispanic)
````
